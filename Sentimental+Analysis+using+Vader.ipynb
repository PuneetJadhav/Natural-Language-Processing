#Sentiment Analysis done using Vader is one of the best currently available.
#Vader package in nltk not only catches the key words from lexicon but it also catches expression and exclamations like :) or :( and so on.
Vader gives 4 types of output for each text. % of positive,negative and neutral and the most important output which is combination of all the other 3 types compund .Compound value is between -1 and 1 . Negative compound means the sentence is negative and as the value is more and more towards 1 we can predict the sentence to be more positive.

Below is the comparision for 1000 reviews from 1 restaurant done by customers. 
The actual review and vader review is shown in dataframe at the end.




```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sentiment=SentimentIntensityAnalyzer()
```

```python
import numpy as np
import pandas as pd
import os
import warnings
import nltk
warnings.filterwarnings('ignore')
os.chdir('C:\\College\\Machine Learnings\\Part 7 - Natural Language Processing')
```

WE now load all the necessary package and download stopwords to create corpus

```python
from nltk.downloader import  *
nltk.download('popular')
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer as ps
dataset=pd.read_csv('Restaurant_Reviews.tsv',delimiter='\t',quoting=3)
```

```python
score=[]
reviews_dataframe=pd.DataFrame()
from pandas import DataFrame,Series
row_number=0
for review in dataset.iloc[:,0]:
    column_number=0
    review_dict=sentiment.polarity_scores(text=review)
    reviews_dataframe.loc[row_number,column_number]=review
    column_number=column_number+1
    reviews_dataframe.loc[row_number,column_number]=review_dict['pos']
    column_number=column_number+1
    reviews_dataframe.loc[row_number,column_number]=review_dict['neg']
    column_number=column_number+1
    reviews_dataframe.loc[row_number,column_number]=review_dict['neu']
    column_number=column_number+1
    reviews_dataframe.loc[row_number,column_number]=review_dict['compound']
    row_number=row_number+1
```

```python
reviews_dataframe.columns=['Review','Positive',"Negative","Neutral","Compound"]
```

```python
actual_reviews=Series(dataset.iloc[:,1])
```

```python
reviews_dataframe=pd.concat(objs=[reviews_dataframe,actual_reviews],axis=1)
```

```python
reviews_dataframe.rename(columns={"Liked":"Actual Reviews"},inplace=True)
```
